```python
import hmmlearn
import hmmlearn.hmm
import numpy as np
import nltk
```

```python
text_file = open('C:/disk_D/1_tampere_uni/text data analysis/hw/8/exercises08_09/hmm_sentences.txt')
sentences = text_file.readlines()

for i in range(len(sentences)):
    sentences[i] = nltk.Text(sentences[i].split())
```

```python
vocabularies = [] 
indices_in_vocabularies = []
# Find the vocabulary of each document 
for document in sentences:   
    # Get unique words and where they occur       
    uniqueresults = np.unique(document,return_inverse=True)    
    uniquewords = uniqueresults[0]    
    wordindices = uniqueresults[1]    
    # Store the vocabulary and indices of document words in it    
    vocabularies.append(uniquewords)    
    indices_in_vocabularies.append(wordindices) 
    
tempvocabulary = []  
for document in sentences:   
    tempvocabulary.extend(document) 
    # Find the unique elements among all vocabularies 
uniqueresults = np.unique(tempvocabulary,return_inverse=True) 
unifiedvocabulary = uniqueresults[0] 
wordindices = uniqueresults[1] 
# Translate previous indices to the unified vocabulary. 
# Must keep track where each vocabulary started in 
# the concatenated one. 
vocabularystart = 0 
indices_in_unifiedvocabulary = [] 
for k in range(len(sentences)):    
    # In order to shift word indices, we must temporarily    
    # change their data type to a Numpy array    
    tempindices = np.array(indices_in_vocabularies[k])    
    tempindices = tempindices + vocabularystart    
    tempindices = wordindices[tempindices]    
    indices_in_unifiedvocabulary.append(tempindices)    
    vocabularystart = vocabularystart + len(vocabularies[k])
```

```python
unifiedvocabulary
```

```python
concatenated_data=[] 
documentlengths=[] 

for k in range(len(indices_in_unifiedvocabulary)):    
    concatenated_data.extend(indices_in_unifiedvocabulary[k])    
    documentlengths.append(len(indices_in_unifiedvocabulary[k]))
    
concatenated_data = np.matrix(concatenated_data).T 
```

```python
model = hmmlearn.hmm.CategoricalHMM(n_components=10, n_iter=100, verbose=True) 
model_fitted = model.fit(concatenated_data, lengths=documentlengths)
```

```python
print('start probabilities')
print(model_fitted.startprob_)
print('emission probabilities')
print(model_fitted.emissionprob_)
print('transition probabilities')
print(model_fitted.transmat_)
```

```python
for idx, ep_b in enumerate(model_fitted.emissionprob_):
    print('Beta_', (idx+1), ':', end='')
    for n, pr in enumerate(ep_b):
        if n % 11 == 0:
            print(' \\\\ ')
        print(round(pr,4), end=',')
    print('; \\\\')
```

```python
for idx, tr_p in enumerate(model_fitted.transmat_):
    print('$theta_', (idx+1), '$: ', end='')
    for n, p in enumerate(tr_p):
        if n == 9:
            print(round(p,5), end='')
        else:   
            print(round(p,5), end=', ')
    print('; \\\\')
```
