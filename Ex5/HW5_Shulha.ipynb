```python
import requests
import bs4
import numpy as np
import nltk
import matplotlib.pyplot as plt
import nltk.lm
import scipy
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger') 
nltk.download('wordnet') 
nltk.download('stopwords')
nltk.download('omw-1.4')
lemmatizer=nltk.stem.WordNetLemmatizer() 
```

```python
def remove_header_footer(text, name_book):
    header = "*** START OF THIS PROJECT GUTENBERG EBOOK " + name_book.upper() + " ***"
    footer = "*** END OF THIS PROJECT GUTENBERG EBOOK " + name_book.upper() + " ***"
    no_header_text = text[(text.find(header) + len(header)):]
    no_footer_text = no_header_text[:no_header_text.find(footer)]
    
    return no_footer_text

def tokenize_text(text):
    
    temp_tokenizedtext = nltk.word_tokenize(text)    
    mycrawled_nltktext = nltk.Text(temp_tokenizedtext)    
        
    return mycrawled_nltktext


def lower_case_text(text):
    mycrawled_lowercasetext = [] 

    for k in range(len(text)):        
        lowercaseword = text[k].lower()        
        mycrawled_lowercasetext.append(lowercaseword)    
    
    return mycrawled_lowercasetext

def tagtowordnet(postag):   
    wordnettag = -1   
    if postag[0] == 'N':        
        wordnettag = 'n'   
    elif postag[0] == 'V':        
        wordnettag = 'v'   
    elif postag[0] == 'J':        
        wordnettag = 'a'    
    elif postag[0] == 'R':        
        wordnettag = 'r'    
    return(wordnettag)

def lemmatizetext(nltktexttolemmatize):    
    # Tag the text with POS tags    
    taggedtext = nltk.pos_tag(nltktexttolemmatize)   
    # Lemmatize each word text    
    lemmatizedtext = []    
    for l in range(len(taggedtext)):       
        # Lemmatize a word using the WordNet converted POS tag       
        wordtolemmatize = taggedtext[l][0]        
        wordnettag = tagtowordnet(taggedtext[l][1])        
        if wordnettag != -1:            
            lemmatizedword = lemmatizer.lemmatize(wordtolemmatize,wordnettag)        
        else:            
            lemmatizedword=wordtolemmatize       
            # Store the lemmatized word        
        lemmatizedtext.append(lemmatizedword)
        
    return(lemmatizedtext) 


def make_vocabulary(text):
    myvocabulary = [] 
    myindices_in_vocabulary = []
    # Find the vocabulary of each document    
    # Get unique words and where they occur      
    uniqueresults = np.unique(text,return_inverse=True)   
     # Store the vocabulary and indices
    myvocabulary = uniqueresults[0]    
    myindices_in_vocabulary = uniqueresults[1]    
    
    return myvocabulary, myindices_in_vocabulary
```

```python
# get a content of the book, remove header and footer text, replace \n\r by space
book_merry_html = requests.get('https://www.gutenberg.org/files/964/964.txt')
book_merry_text = bs4.BeautifulSoup(book_merry_html.content,'html.parser')
book_merry_text = remove_header_footer(str(book_merry_text), 'THE MERRY ADVENTURES OF ROBIN HOOD')
book_merry_text = ' '.join(str(book_merry_text).split())

# tokenizing, lower casing and creating the vocabulary of the book with indices
tokenized_book_merry = tokenize_text(book_merry_text)
lower_tokenized_book_merry = lower_case_text(tokenized_book_merry)

book_merry_vocabulary, indices_book_merry_vocabulary = make_vocabulary(lower_tokenized_book_merry)
```

```python
lower_tokenized_book_merry_sentences = []
temp_sentence = []
for word in lower_tokenized_book_merry:
    if word == '.' or word == '?' or word == '!' or word == '...':
        lower_tokenized_book_merry_sentences.append(temp_sentence)
        temp_sentence = []
        continue
    temp_sentence.append(word)
```

```python
maxN = [1, 2, 3, 5]

for N in maxN:
    print(N, '-gram model')
    ngram_train_book_merry, padded_sentences_book_merry =\
                                             nltk.lm.preprocessing.padded_everygram_pipeline(N, 
                                                                                             lower_tokenized_book_merry_sentences)
    
    ngramm_model = nltk.lm.MLE(N)
    ngramm_model.fit(ngram_train_book_merry, padded_sentences_book_merry)
    
    gen_text1 = ['</s>']
    while '</s>' in gen_text1:
        gen_text1 = ngramm_model.generate(num_words=100)
        
    for word in gen_text1:
        print(word, end=' ')
    print()
    
    gen_text2 = ['</s>']
    while '</s>' in gen_text2:
        gen_text2 = ngramm_model.generate(num_words=120)
    
    for word in gen_text2:
        print(word, end=' ')
    print()
    
    if N in [2, 3, 5]:
        gen_text3 = ['</s>']
        while '</s>' in gen_text3:
            gen_text3 = ngramm_model.generate(num_words=100, text_seed=['the', 'moon'])
        print('the moon ', end='')
        for word in gen_text3:
            print(word, end=' ')
    print()
```
